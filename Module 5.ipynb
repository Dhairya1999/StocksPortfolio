### Query 2:

```python
d1 = pd.read_csv('TCS.csv')
d2 = pd.read_csv('RELIANCE.csv')
d3 = pd.read_csv('FORTIS.csv')
d4 = pd.read_csv('TATAPOWER.csv')
d5 = pd.read_csv('INFY.csv')
d1 = d1[d1.Series == 'EQ']
d2 = d2[d2.Series == 'EQ']
d3 = d3[d3.Series == 'EQ']
d4 = d4[d4.Series == 'EQ']
d5 = d5[d5.Series == 'EQ']
d1.dropna(inplace = True)
d2.dropna(inplace = True)
d3.dropna(inplace = True)
d4.dropna(inplace = True)
d5.dropna(inplace = True)
data = pd.concat([d1['Close Price'],d2['Close Price'],d3['Close Price'],d4['Close Price'],d5['Close Price']],axis=1)
data.columns = ['TCS','REL','FORT','TATA','INFY']
data
```

```python
returns = data.pct_change()
mean_daily_returns = returns.mean()
cov_matrix = returns.cov()

weights = np.asarray([0.20,0.20,0.20,0.20,0.20])

portfolio_return = round(np.sum(mean_daily_returns * weights) *252,2)
portfolio_std_dev = round(np.sqrt(np.dot(weights.T,np.dot(cov_matrix,weights))) * np.sqrt(252),2)

print("The anual mean return on the stock is: ",portfolio_return)
print("The anual volatility of the stock is: ",portfolio_std_dev)
```

### Query 3,4

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(20,10))

num_portfolios = 25000
results = np.zeros((3+data.columns.size,num_portfolios))
for i in range(num_portfolios):
    weights = np.array(np.random.random(data.columns.size))
    weights /= np.sum(weights)

    portfolio_return = np.sum(mean_daily_returns * weights) *252
    portfolio_std_dev = np.sqrt(np.dot(weights.T,np.dot(cov_matrix,weights))) * np.sqrt(252)

    results[0,i] = portfolio_return
    results[1,i] = portfolio_std_dev
    results[2,i] = results[0,i] / results[1,i]
    
    for j in range(len(weights)):
        results[j+3,i] = weights[j]
results_frame = pd.DataFrame(results.T,columns = ['ret','stdev','sharpe','TCS','REL','FORT','TATA','INFY'])
results_frame
    
max_sharpe_port = results_frame.iloc[results_frame['sharpe'].idxmax()]
min_vol_port = results_frame.iloc[results_frame['stdev'].idxmin()]

plt.scatter(results_frame.stdev,results_frame.ret,c=results_frame.sharpe,cmap='RdYlBu')
plt.xlabel('Volatility')
plt.ylabel('Returns')

plt.colorbar()

plt.scatter(max_sharpe_port[1],max_sharpe_port[0],marker=(5,1,0),color='r',s=200)
plt.scatter(min_vol_port[1],min_vol_port[0],marker=(5,1,0),color='g',s=200)
```
