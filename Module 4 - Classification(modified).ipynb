```python
#Making the bollinger bands 
#Considering the 14 day moving average and the 20 standard deviation
#Upper band is two SD above and Lower band two SD below

import pandas as pd
import matplotlib.pyplot as plt

TCS = pd.read_csv('TCS.csv')

TCS['14 day MV'] = TCS['Close Price'].rolling(window=14).mean()
TCS['30 day SD'] = TCS['Close Price'].rolling(window=20).std()
TCS['Upper Band'] = TCS['14 day MV'] + ( TCS['30 day SD'] * 2)
TCS['Lower Band'] = TCS['14 day MV'] - ( TCS['30 day SD'] * 2)
TCS['Middle Band'] = (TCS['Upper Band'] + TCS['Lower Band'])/2

TCS = TCS.dropna()
TCS
```

```python
#'Buy' if the stock price is below the lower Bollinger band
#'Hold Buy/ Liquidate Short' if the stock price is between the lower and middle Bollinger band
#'Hold Short/ Liquidate Buy' if the stock price is between the middle and upper Bollinger band
#'Short' if the stock price is above the upper Bollinger band
# Define a function to add the values.

def select_buy(TCS):
    if TCS['Close Price'] < TCS['Lower Band']:
        return "Buy"
    elif TCS['Close Price'] > TCS['Lower Band'] and TCS['Close Price'] < TCS['Middle Band']:
        return "Hold Buy / Liquidate Short"
    elif TCS['Close Price'] > TCS['Middle Band'] and TCS['Close Price'] < TCS['Upper Band']:
        return "Hold Short / Liquidate Buy"
    elif TCS['Close Price'] > TCS['Upper Band']:
        return "Short"
    
# Make a new column named CALL to enter the returned value from the function

TCS = TCS.assign( Call = TCS.apply(select_buy,axis=1))
TCS
```

```python
# Transforming the model according to our requirnmnents

# Importing the sklearn preprocessing function to transform the data into a favourable format
from sklearn import preprocessing

# Using the labelencoder to assign numeric labels to the features in our call column 
labelencoder  = preprocessing.LabelEncoder()

#creating train data 
Train_X = TCS[['Upper Band','Middle Band','Lower Band','Close Price']]
Train_Y = labelencoder.fit(TCS['Call'])

for i,items in enumerate(Train_Y.classes_):
    print(items,"=>",i)
    
Train_Y = labelencoder.transform(TCS['Call'])

# Reshaping the data according to our dataframe in 479 columns
Train_Y = Train_Y.reshape(-1,1)
```

```python
import sklearn  as sk
from sklearn.linear_model import LogisticRegression

#The solver uses a coordinate descent (CD) algorithm that solves optimization problems. 
#It does so by successively performing approximate minimization along coordinate directions or coordinate hyperplanes.

LR = LogisticRegression(random_state=0,solver='lbfgs',multi_class='ovr')

#Fitting the model
LR.fit(Train_X,Train_Y.ravel())
LR.predict(Train_X)
print("Logistic Regression: ")

Accuracy = LR.score(Train_X,Train_Y) * 100
print("Accuracy: ",Accuracy,"%")
```

```python
import sklearn as sk
from sklearn import svm

# Make an instance of the support vector machine
# Using the linear support vector classifier 
SVM = svm.LinearSVC()
SVM.fit(Train_X,Train_Y.ravel())

SVM.predict(Train_X)
print("Support Vector MAchine: ")

Accuracy = SVM.score(Train_X,Train_Y) * 100
print("Accuracy: ",Accuracy,"%")
```

```python
import sklearn as sk  
from sklearn.ensemble import RandomForestClassifier

#import the random forest classifier from the sklearn library
# Create the instance of random forest 
# n_estimators is the number of decision tree
RF = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
RF.fit(Train_X, Train_Y)

RF.predict(Train_X)
print("Random Forests")

Accuracy = RF.score(Train_X,Train_Y) * 100
print("Accuracy: ",Accuracy,"%")
```
