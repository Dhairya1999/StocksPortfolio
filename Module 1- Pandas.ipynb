### Query 1.5

Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def avg_price(N):
    return (((data['Average Price'].tail(N)).sum())/N)
print("Average price of last N days: ")
print("Average price of last 1 week is: ",avg_price(5))
print("Average price of last 2 week is: ",avg_price(10))
print("Average price of last 1 month is: ",avg_price(20))
print("Average price of last 2 month is: ",avg_price(60))
print("Average price of last 6 month is: ",avg_price(120))
print("Average price of last 1 year is: ",avg_price(240))

```

```python

def profitloss(N):
    d=data['Close Price'].tail(N).iloc[N-1] - data['Close Price'].tail(N).iloc[0]
    if d<0:
        loss = -(d)
        loss_percentage = (loss/(data['Close Price'].tail(N).iloc[N-1]))*100
        return loss_percentage
    else:
        profit = d
        profit_percentage = (profit/(data['Close Price'].tail(N).iloc[N-1]))*100
        return profit_percentage
print("percentage profit/loss of last N days: ")
print("percentage profit/loss last 1 week is: ",profitloss(5),"%")
print("percentage profit/loss 2 week is: ",profitloss(10),"%")
print("percentage profit/loss 1 month is: ",profitloss(20),"%")
print("percentage profit/loss 2 month is: ",profitloss(60),"%")
print("percentage profit/loss 6 month is: ",profitloss(120),"%")
print("percentage profit/loss 1 year is: ",profitloss(240),"%")
```

### Query 1.6

Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. ### Query 1.7

 Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
'Positive' for 'Day_Perc_Change' in between 1 and 3
'Negative' for 'Day_Perc_Change' in between -1 and -3
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
'Among top losers' for 'Day_Perc_Change' in between -3 and -7
'Bull run' for 'Day_Perc_Change' >7
'Bear drop' for 'Day_Perc_Change' <-7

```python
data['Trend']=0
for i in np.arange(data.Day_Perc_Change.size):
    if ((data['Day_Perc_Change'][i] >= -0.5) & (data['Day_Perc_Change'][i] <= 0.5)):
        data['Trend'][i] = 'Slight or No change'
    elif ((data['Day_Perc_Change'][i] >= 0.5) & (data['Day_Perc_Change'][i] <= 1)):
        data['Trend'][i] = 'Slight positive'
    elif ((data['Day_Perc_Change'][i] <= -0.5) & (data['Day_Perc_Change'][i] >= -1)):
        data['Trend'][i] = 'Slight negative'
    elif ((data['Day_Perc_Change'][i] >= 1) & (data['Day_Perc_Change'][i] <= 3)):
        data['Trend'][i] = 'Positive' 
    elif ((data['Day_Perc_Change'][i] <= -1) & (data['Day_Perc_Change'][i] >= -3)):
        data['Trend'][i] = 'Negative'
    elif ((data['Day_Perc_Change'][i] >= 3) & (data['Day_Perc_Change'][i] <= 7)):
        data['Trend'][i] = 'Among top gainers'
    elif ((data['Day_Perc_Change'][i] <= -3) & (data['Day_Perc_Change'][i] >= -7)):
        data['Trend'][i] = 'Among top losers'
    elif (data['Day_Perc_Change'][i] > 7):
        data['Trend'][i] = 'Bull run' 
    elif (data['Day_Perc_Change'][i] < -7):
        data['Trend'][i] = 'Bear drop' 
data
```
